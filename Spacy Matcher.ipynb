```python
with open("wiki_mlk.txt","r") as f:
    text=f.read()
```

```python
text
```

```python
matcher=Matcher(nlp.vocab)
pattern=[{"POS":"PROPN"}]   # part of speech
matcher.add("PROPER_NOUN",[pattern])
doc=nlp(text)
matches=matcher(doc)
print(len(matches))
for match in matches[:10]:
    print(match, doc[match[1]:match[2]])
```

```python
matcher=Matcher(nlp.vocab)
pattern=[{"POS":"PROPN","OP":"+"}]   # part of speech 
matcher.add("PROPER_NOUN",[pattern])
doc=nlp(text)
matches=matcher(doc)
print(len(matches))
for match in matches[:10]:
    print(match, doc[match[1]:match[2]])
```

```python
matcher=Matcher(nlp.vocab)
pattern=[{"POS":"PROPN","OP":"+"}]   # part of speech
matcher.add("PROPER_NOUN",[pattern],greedy='LONGEST')
doc=nlp(text)
matches=matcher(doc)
print(len(matches))
for match in matches[:10]:
    print(match, doc[match[1]:match[2]])
```

```python
matcher=Matcher(nlp.vocab)
pattern=[{"POS":"PROPN","OP":"+"}]   # part of speech(POS) proper noun
matcher.add("PROPER_NOUN",[pattern],greedy='LONGEST')
doc=nlp(text)
matches=matcher(doc)
matches.sort(key=lambda x:x[1])  # sort
print(len(matches))
for match in matches[:10]:
    print(match, doc[match[1]:match[2]]) # we r geting proper noun with sequentially
```

```python
matcher=Matcher(nlp.vocab)
pattern=[{"POS":"PROPN","OP":"+"},{"POS":"VERB"}]   # part of speech(POS) proper noun + Verb
matcher.add("PROPER_NOUN",[pattern],greedy='LONGEST')
doc=nlp(text)
matches=matcher(doc)
matches.sort(key=lambda x:x[1])  # sort
print(len(matches))
for match in matches[:10]:
    print(match, doc[match[1]:match[2]]) # we r geting proper noun with sequentially
```

```python
import json
with open('alice.json','r') as f:
    data=json.load(f)
```

```python
text=data[0][2][0]
```

```python
text
```

```python
# standardised text as much as possible
text=text.replace("`","'")
text
```

```python
matcher=Matcher(nlp.vocab)
pattern=[{"ORTH":"'"}, # start with '
        {"IS_ALPHA":True,"OP":"+"}, # alphabets
        {"IS_PUNCT":True,"OP":"*"}, # like ,
        {"ORTH":"'"}]   # end with '
matcher.add("PROPER_NOUN",[pattern],greedy='LONGEST')
doc=nlp(text)
matches=matcher(doc)
matches.sort(key=lambda x:x[1])  # sort
print(len(matches))
for match in matches[:10]:
    print(match, doc[match[1]:match[2]])
```

```python
speak_lemmas=['think',"say"] # thought
matcher=Matcher(nlp.vocab)
pattern=[{"ORTH":"'"}, # start with '
        {"IS_ALPHA":True,"OP":"+"}, # alphabets
        {"IS_PUNCT":True,"OP":"*"}, # like ,
        {"ORTH":"'"}, # end with '
        {"POS":"VERB","LEMMA":{"IN":speak_lemmas}},
        {"POS":"PROPN","OP":"+"}, # alice
        {"ORTH":"'"}, # start with '
        {"IS_ALPHA":True,"OP":"+"}, # alphabets
        {"IS_PUNCT":True,"OP":"*"}, # like ,
        {"ORTH":"'"} # end with '
        ] 
matcher.add("PROPER_NOUN",[pattern],greedy='LONGEST')
doc=nlp(text)
matches=matcher(doc)
matches.sort(key=lambda x:x[1])  # sort
print(len(matches))
for match in matches[:10]:
    print(match, doc[match[1]:match[2]])  # the pattern is singular not multi
```

```python

speak_lemmas = ["think", "say"]
text = data[0][2][0].replace( "`", "'")
matcher=Matcher(nlp.vocab)
pattern1 = [{'ORTH': "'"}, {'IS_ALPHA': True, "OP": "+"}, {'IS_PUNCT': True, "OP": "*"}, {'ORTH': "'"}, {"POS": "VERB", "LEMMA": {"IN": speak_lemmas}}, {"POS": "PROPN", "OP": "+"}, {'ORTH': "'"}, {'IS_ALPHA': True, "OP": "+"}, {'IS_PUNCT': True, "OP": "*"}, {'ORTH': "'"}]
pattern2 = [{'ORTH': "'"}, {'IS_ALPHA': True, "OP": "+"}, {'IS_PUNCT': True, "OP": "*"}, {'ORTH': "'"}, {"POS": "VERB", "LEMMA": {"IN": speak_lemmas}}, {"POS": "PROPN", "OP": "+"}]
pattern3 = [{"POS": "PROPN", "OP": "+"},{"POS": "VERB", "LEMMA": {"IN": speak_lemmas}}, {'ORTH': "'"}, {'IS_ALPHA': True, "OP": "+"}, {'IS_PUNCT': True, "OP": "*"}, {'ORTH': "'"}]
matcher.add("PROPER_NOUNS", [pattern1, pattern2, pattern3], greedy='LONGEST')
for text in data[0][2]:
    text = text.replace("`", "'")
    doc = nlp(text)
    matches = matcher(doc)
    matches.sort(key = lambda x: x[1])
    print (len(matches))
    for match in matches[:10]:
        print (match, doc[match[1]:match[2]])  # multi pattern
```
