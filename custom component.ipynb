```python
import spacy
```

```python
nlp=spacy.load("en_core_web_sm")
doc=nlp("Britain is a place. Mary is a doctor.")
```

```python
for ent in doc.ents: # GPE geopolitical entity
    print(ent.text, ent.label_)
```

```python
from spacy import Language
```

```python
@Language.component("remove_gpe")  # custom component

def remove_gpe(doc):
    original_ents=list(doc.ents)
    for ent in doc.ents:
        if ent.label_=="GPE":
            original_ents.remove(ent)
    doc.ents=original_ents
    return (doc)
```

```python
nlp.add_pipe("remove_gpe")
```

```python
nlp.analyze_pipes()
```

```python

doc=nlp("Britain is a place. Mary is a doctor.")
for ent in doc.ents: # GPE geopolitical entity
    print(ent.text, ent.label_)
```
